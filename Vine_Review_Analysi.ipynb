```python
# Add the Pandas dependency.
import pandas as pd
import numpy as np
```

```python
# Files to load
vine_data = "vine_table.csv"
```

```python
# Read the data file and store it in a Pandas DataFrame.
vine_df = pd.read_csv(vine_data)
vine_df.head(5)
# vine_df.count
```

```python
vine_df_gt = vine_df[vine_df['total_votes'] >= 20]
vine_df_gt.head(5)
```

```python
# vine_df_gt.dtypes
# vine_df_gt = vine_df_gt.assign(hratio = lambda x: (x['helpful_votes'] / x['total_votes'])) 
vine_df_gt['hratio'] = vine_df_gt['helpful_votes'] / vine_df_gt['total_votes']
vine_df_hf = vine_df_gt[vine_df_gt['hratio'] >= .5]
vine_df_hf.head(5)
```

```python
vine_df_p = vine_df_hf[vine_df_hf['vine'] == 'Y']
vine_df_p.head(5)
```

```python
vine_df_up = vine_df_hf[vine_df_hf['vine'] == 'N']
vine_df_up.head(5)
```

```python
#paid dataframe
#number of rows
np = vine_df_p.shape[0]
# print('Number of Reviews - Vine Program :',num_rows)
print('Number of Reviews - Vine Program :',len(vine_df_p))
star5 = vine_df_p.groupby(by='star_rating').agg('count')
# print(star5)
print('Number of 5 Star Reviews - Vine Program :', star5.iloc[3]['review_id'])
pct5 = (star5 / np * 100).round(2)
# print(pct5)
print('Percentage of 5 Star Reviews - Vine Program :', pct5.iloc[3]['review_id'])
```

```python
#unpaid dataframe
#number of rows
npu = vine_df_up.shape[0]
# print('Number of Reviews - Not Vine Program :',num_rows)
print('Number of Reviews - Not Vine Program :',len(vine_df_up))
star5u = vine_df_up.groupby(by='star_rating').agg('count')
# print(star5u)
print('Number of 5 Star Reviews - Not Vine Program :', star5u.iloc[4]['review_id'])
pct5u = (star5u / npu * 100).round(2)
# print(pct5u)
print('Percentage of 5 Star Reviews - Not Vine Program :', pct5u.iloc[4]['review_id'])
```
